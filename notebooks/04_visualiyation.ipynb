```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("results/cell_clusters.csv")

sns.scatterplot(x="area", y="circularity", hue="cluster", data=df, palette="Set2")
plt.title("Cell Morphology Clusters")
plt.savefig("results/plots/cluster_plot.png")
plt.show()
```
