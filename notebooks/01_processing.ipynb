```python
import sys
import os
sys.path.append(os.path.abspath("../src"))
import cv2
import os
import matplotlib.pyplot as plt
from skimage.filters import gaussian
from skimage import io
from utils import generate_synthetic_cells
import numpy as np

generate_synthetic_cells()

input_dir = "data/synthetic_cells"
output_dir = "results/plots"
os.makedirs(output_dir, exist_ok=True)

def preprocess(img_path):
    img = io.imread(img_path)
    blurred = gaussian(img, sigma=1)
    _, thresh = cv2.threshold((blurred * 255).astype(np.uint8), 50, 255, cv2.THRESH_BINARY)
    return thresh

for fname in os.listdir(input_dir):
    path = os.path.join(input_dir, fname)
    mask = preprocess(path)
    plt.imsave(f"{output_dir}/{fname}", mask, cmap="gray")
```
