This plot shows how cell area and circularity relate before dimensionality reduction. 
It offers a biologically intuitive view of how shape and size vary across clusters.

While PCA provides a compressed representation of feature variance, this raw feature plot helps validate that clustering reflects meaningful morphological differences â€” such as irregular shape or size extremes.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

df = pd.read_csv("results/cell_features.csv")
X = df[["area", "perimeter", "circularity"]]
X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

kmeans = KMeans(n_clusters=3, random_state=42)
df["cluster"] = kmeans.fit_predict(X_pca)
df.to_csv("results/cell_clusters.csv", index=False)
```
